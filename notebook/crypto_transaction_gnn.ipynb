```python
import pandas as pd 
import numpy as np
```

```python
features=pd.read_csv('elliptic_txs_features.csv')
edges=pd.read_csv('elliptic_txs_edgelist.csv')
classes=pd.read_csv('elliptic_txs_classes.csv')
```

```python
features.columns=['txId','time']+[f'feat_{i}' for i in range(features.shape[1]-2)]
classes.columns=['txId','class']
data=features.merge(classes,on='txId')
print("merged shape:",data.shape)
data.head()
```

```python
data=data[data['class']!='unknown']
print("after removing unknown:",data.shape)
data['class'].value_counts()
```

```python
type(data)
```

```python
valid_txIds = set(data['txId'])
edges_filtered = edges[
    (edges['txId1'].isin(valid_txIds)) &
    (edges['txId2'].isin(valid_txIds))
]

print("Filtered edges shape:", edges_filtered.shape)
edges_filtered.head()
```

```python
txId_to_index = {txId: idx for idx, txId in enumerate(data['txId'])}
print("Total nodes:", len(txId_to_index)) #convert big id to small numbers
```

```python
edge_index = edges_filtered.apply(
    lambda row: [txId_to_index[row['txId1']], txId_to_index[row['txId2']]],
    axis=1
)

edge_index = np.array(edge_index.tolist()).T

print("Edge index shape:", edge_index.shape)
```

```python
import torch
x=data.drop(columns=['txId','class'])
x=torch.tensor(x.values,dtype=torch.float)
print("node feature shape:",x.shape)
```

```python

y = data['class'].astype(int)

y = y.replace({2: 0, 1: 1})

y = torch.tensor(y.values, dtype=torch.long)

print("Label shape:", y.shape)
```

```python
edge_index = torch.tensor(edge_index, dtype=torch.long)

print("Edge index tensor shape:", edge_index.shape)
```

```python
from torch_geometric.data import Data

graph_data = Data(x=x, edge_index=edge_index, y=y)

print(graph_data)
```

```python
import torch
import numpy as np

num_nodes = graph_data.num_nodes
indices = np.random.permutation(num_nodes)

train_size = int(0.8 * num_nodes)

train_idx = indices[:train_size]
test_idx = indices[train_size:]

train_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask = torch.zeros(num_nodes, dtype=torch.bool)

train_mask[train_idx] = True
test_mask[test_idx] = True

graph_data.train_mask = train_mask
graph_data.test_mask = test_mask

print("Train nodes:", train_mask.sum().item())
print("Test nodes:", test_mask.sum().item())
```

```python
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

class GCN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GCN, self).__init__()
        
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, training=self.training)
        x = self.conv2(x, edge_index)
        
        return x
```

```python
input_dim = graph_data.num_node_features
hidden_dim = 64
output_dim = 2

model = GCN(input_dim, hidden_dim, output_dim)

print(model)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.CrossEntropyLoss()
```

```python
def train():
    model.train()
    optimizer.zero_grad()
    
    out = model(graph_data)
    
    loss = criterion(out[graph_data.train_mask], 
                     graph_data.y[graph_data.train_mask])
    
    loss.backward()
    optimizer.step()
    
    return loss.item()


def test():
    model.eval()
    out = model(graph_data)
    pred = out.argmax(dim=1)
    
    correct = pred[graph_data.test_mask] == graph_data.y[graph_data.test_mask]
    acc = int(correct.sum()) / int(graph_data.test_mask.sum())
    
    return acc


for epoch in range(1, 51):
    loss = train()
    acc = test()
    
    if epoch % 5 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}, Test Acc: {acc:.4f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

model.eval()
out = model(graph_data)
pred = out.argmax(dim=1)

y_true = graph_data.y[graph_data.test_mask].cpu().numpy()
y_pred = pred[graph_data.test_mask].cpu().numpy()

print(confusion_matrix(y_true, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred))
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

classes = np.unique(y_true)
weights = compute_class_weight(class_weight='balanced',
                               classes=classes,
                               y=y_true)

class_weights = torch.tensor(weights, dtype=torch.float)

print(class_weights)
```

```python
criterion = nn.CrossEntropyLoss(weight=class_weights)
```

```python

model = GCN(input_dim, hidden_dim, output_dim)


optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

criterion = nn.CrossEntropyLoss(weight=class_weights)
```

```python
for epoch in range(1, 51):
    loss = train()
    acc = test()
    
    if epoch % 5 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}, Test Acc: {acc:.4f}")
```

```python
model.eval()
out = model(graph_data)
pred = out.argmax(dim=1)

y_true = graph_data.y[graph_data.test_mask].cpu().numpy()
y_pred = pred[graph_data.test_mask].cpu().numpy()

print(confusion_matrix(y_true, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred))
```

```python
from torch_geometric.nn import SAGEConv
import torch.nn.functional as F
import torch.nn as nn

class GraphSAGE(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GraphSAGE, self).__init__()
        
        self.conv1 = SAGEConv(input_dim, hidden_dim)
        self.conv2 = SAGEConv(hidden_dim, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        
        return x
```

```python
input_dim = graph_data.num_node_features
hidden_dim = 64
output_dim = 2

model = GraphSAGE(input_dim, hidden_dim, output_dim)

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.CrossEntropyLoss(weight=class_weights)

print(model)
```

```python
for epoch in range(1, 51):
    loss = train()
    acc = test()
    
    if epoch % 5 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}, Test Acc: {acc:.4f}")
```

```python
model.eval()
out = model(graph_data)
pred = out.argmax(dim=1)

y_true = graph_data.y[graph_data.test_mask].cpu().numpy()
y_pred = pred[graph_data.test_mask].cpu().numpy()

print(confusion_matrix(y_true, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

model.eval()

with torch.no_grad():
    out = model(graph_data)
    probs = torch.softmax(out, dim=1)[:, 1] 

y_true = graph_data.y[graph_data.test_mask].cpu().numpy()
y_scores = probs[graph_data.test_mask].cpu().numpy()
```

```python
auc = roc_auc_score(y_true, y_scores)
print("ROC-AUC Score:", auc)
```

```python
fpr, tpr, thresholds = roc_curve(y_true, y_scores)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f"GraphSAGE (AUC = {auc:.3f})")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
pr_auc = average_precision_score(y_true, y_scores)

print("PR-AUC Score:", pr_auc)

plt.figure(figsize=(6,6))
plt.plot(recall, precision, label=f"GraphSAGE (AP = {pr_auc:.3f})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```
